```
##import wave and load audio

import wave
import numpy as np

def frame_rate(audio_path):
  fin = wave.open(audio_path, 'rb')
  sample_rate = fin.getframerate()
  fin.close()
  return sample_rate

def load_audio(audio_path):
  fin = wave.open(audio_path, 'rb')
  audio = np.frombuffer(fin.readframes(fin.getnframes()), np.int16)
  fin.close()
  return audio
```
